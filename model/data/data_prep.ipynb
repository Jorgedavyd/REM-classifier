Available at the next link. We have to do some data preprocessing so it is legible by the models.

```python
import pandas as pd
import os
```

```python
#Getting the individuals
path = 'outputs/features'

def get_individuals():
    individuals = set()
    for filename in os.listdir(path):
        if '.out' in filename:
            file_path = os.path.join(path, filename)
            individuals.add(file_path.split('_')[0].split('/')[2])
    return list(individuals)

individuals = get_individuals()
```

```python
#Getting the data from the .out files
features = ['count', 'hr', 'cosine', 'time', 'psg']
def data_(individual):
    df_list = []
    for filename in os.listdir(path):
        file_path = os.path.join(path, filename)
        if os.path.isfile(file_path) and individual in file_path:
            feature = file_path.split('_')[1]
            with open(file_path, 'r') as file:
                lines = file.readlines()
            data = []

            for line in lines:
                line = float(line[:-1])
                data.append(line)
            df = pd.DataFrame(data, columns = [feature])
            df_list.append(df)
    
    database = pd.concat(df_list, axis = 1)
    database = database.reindex(columns = features)

    return database
```

```python
#Concatenate all individual dataframes
database_list = []
for i in get_individuals():
    database_list.append(data_(i))

df = pd.concat(database_list,keys = individuals)
```

```python
#REM and NREM classification
df.loc[df['psg']!=5,['psg']] = 0
df.loc[df['psg']==5,['psg']] = 1
df
```
